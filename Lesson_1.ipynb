```python
def show_ell(a, b):
    x = np.linspace(-a, a, 100)

    y1 = [math.sqrt(b**2 * (1 - xi**2 / a**2)) for xi in x]
    y2 = [- math.sqrt(b**2 * (1 - xi**2 / a**2)) for xi in x]
    
    plt.plot(x, y1)
    plt.plot(x, y2)

    plt.show()
```

```python
def show_circle(r):
    x = np.linspace(-r, r, 100)
    

    y1 = [math.sqrt(r**2 - xi**2) for xi in x]
    y2 = [- math.sqrt(r**2 - xi**2) for xi in x]
    
    plt.plot(x, y1)
    plt.plot(x, y2)

    plt.show()
```

```python
def show_hip(a, b):
    x = np.hstack([np.linspace(-2 * a, -a, 100), np.linspace(a, 2 * a, 100)])
    

    y1 = [math.sqrt(b**2 * (xi**2 / a**2 - 1)) for xi in x]
    y2 = [- math.sqrt(b**2 * (xi**2 / a**2 - 1)) for xi in x]
    
    plt.scatter(x, y1)
    plt.scatter(x, y2)

    plt.show()
```

```python
show_circle(10)
```

```python
show_ell(2, 4)
```

```python
show_hip(1, 2)
```

```python
def get_v_len(v):
    return math.sqrt(sum([vi**2 for vi in v]))
```

```python
x = [20, 20, 0]
get_v_len(x)
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)

X = np.arange(-15, 15, 0.5)
Y = np.arange(-15, 15, 0.5)

X, Y = np.meshgrid(X, Y)

Z1 = 2 * X + 5 * Y
Z2 = 2 * X + 5 * Y - 50
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)

plt.show()
```

```python
fig = figure()
ax = Axes3D(fig)

X = np.arange(-15, 15, 0.5)
Y = np.arange(-15, 15, 0.5)

X, Y = np.meshgrid(X, Y)

Z3 = X**2 - Y**2

ax.plot_surface(X, Y, Z3)

plt.show()
```
